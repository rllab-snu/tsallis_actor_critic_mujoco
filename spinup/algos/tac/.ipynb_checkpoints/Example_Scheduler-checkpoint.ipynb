```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
class Weight:
    def __init__(self, weight_start=1.0, weight_end=0.0, max_iter=200, schedule='constant'):
        self.weight_start = weight_start
        self.weight_end = weight_end
        self.schedule = schedule
        self.max_iter = max_iter
        self.count = 0
        
    def __call__(self, ret=None):
        if self.schedule == 'constant':
            return self.weight_start
        elif self.schedule == 'linear':
            self.count += 1
            if self.count <= self.max_iter*0.2:
                return self.weight_start
            elif self.count >= self.max_iter*0.8:                
                return self.weight_end
            else:
                return self.weight_end-(self.count/self.max_iter-0.8)*(self.weight_start-self.weight_end)/0.6

class Alpha:
    def __init__(self, alpha_start=0.2, alpha_end=1e-2, max_iter=200, speed=0.1, schedule='constant'):
        self.alpha_start = alpha_start
        self.alpha_end = alpha_end
        self.schedule = schedule
        self.max_iter = max_iter
        self.speed = speed
        self.count = 0
        
    def __call__(self, ret=None):
        if self.schedule == 'constant':
            return self.alpha_start
        elif self.schedule == 'logarithm':
            self.count += 1
            if self.count <= self.max_iter*0.2:
                return self.alpha_start
            else:
                return np.maximum(
                    self.alpha_end, 
                    np.minimum(
                        self.alpha_start,
                        self.alpha_start/np.maximum(np.log(self.speed*(self.count-0.2*self.max_iter)+1e-3),1)
                    )
                )

class EntropicIndex:
    def __init__(self, q_start=1.0, q_end=1.0, max_iter=200, schedule='constant'):
        self.q_start = q_start
        self.q_end = q_end
        self.schedule = schedule
        self.max_iter = max_iter
        self.count = 0
        
    def __call__(self, ret=None):
        if self.schedule == 'constant':
            return self.q_end
        elif self.schedule == 'linear':
            self.count += 1
            if self.count <= self.max_iter*0.2:
                return self.q_start
            elif self.count >= self.max_iter*0.8:                
                return self.q_end
            else:
                return self.q_end-(self.count/self.max_iter-0.8)*(self.q_start-self.q_end)/0.6
```

```python
weight = Weight(schedule='constant')
weight_list = []
for _ in range(200):
    weight_list.append(weight())
plt.plot(weight_list,label='Constant')
    
weight = Weight(schedule='linear')
weight_list = []
for _ in range(200):
    weight_list.append(weight())
plt.plot(weight_list,label='Linear')
plt.legend()
plt.show()
```

```python
alpha = Alpha(schedule='constant')
alpha_list = []
for _ in range(200):
    alpha_list.append(alpha())
plt.plot(alpha_list,label='Constant')
    
alpha = Alpha(schedule='logarithm',speed=10.0)
alpha_list = []
for _ in range(200):
    alpha_list.append(alpha())
plt.plot(alpha_list,label='Logarithm (10.0)')

alpha = Alpha(schedule='logarithm',speed=1.0)
alpha_list = []
for _ in range(200):
    alpha_list.append(alpha())
plt.plot(alpha_list,label='Logarithm (1.0)')

alpha = Alpha(schedule='logarithm',speed=0.1)
alpha_list = []
for _ in range(200):
    alpha_list.append(alpha())
plt.plot(alpha_list,label='Logarithm (0.1)')

alpha = Alpha(schedule='logarithm',speed=0.05)
alpha_list = []
for _ in range(200):
    alpha_list.append(alpha())
plt.plot(alpha_list,label='Logarithm (0.05)')

plt.legend()
# plt.ylim([0,0.3])
plt.show()
```

```python
entropicindex = EntropicIndex(schedule='constant')
entropicindex_list = []
for _ in range(200):
    entropicindex_list.append(entropicindex())
plt.plot(entropicindex_list,label='Constant')

entropicindex = EntropicIndex(schedule='linear',q_end=1.5)
entropicindex_list = []
for _ in range(200):
    entropicindex_list.append(entropicindex())
plt.plot(entropicindex_list,label='Linear (q=1.5)')

entropicindex = EntropicIndex(schedule='linear',q_end=2.0)
entropicindex_list = []
for _ in range(200):
    entropicindex_list.append(entropicindex())
plt.plot(entropicindex_list,label='Linear (q=2.0)')

entropicindex = EntropicIndex(schedule='linear',q_end=2.5)
entropicindex_list = []
for _ in range(200):
    entropicindex_list.append(entropicindex())
plt.plot(entropicindex_list,label='Linear (q=2.5)')

entropicindex = EntropicIndex(schedule='linear',q_end=3.0)
entropicindex_list = []
for _ in range(200):
    entropicindex_list.append(entropicindex())
plt.plot(entropicindex_list,label='Linear (q=3.0)')

entropicindex = EntropicIndex(schedule='linear',q_end=5.0)
entropicindex_list = []
for _ in range(200):
    entropicindex_list.append(entropicindex())
plt.plot(entropicindex_list,label='Linear (q=5.0)')

plt.legend()
# plt.ylim([0.7,3.3])
plt.show()
```
